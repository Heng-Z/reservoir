```python
inputs = np.zeros(3500)
inputs[400:420] = 1
inputs[550:570] = 1
inputs[650:670] = 2
inputs[810:830] = 2
inputs[940:960] = 1
inputs[1030:1050] = 2
inputs[1050:1070] = 2
inputs[1790:1810] = 1
inputs[1970:1990] = 2
inputs[2100:2120] = 1
inputs[2200:2220] = 2
inputs[2230:2250] = 1
inputs[2340:2360] = 2
inputs[2450:2470] = 1
inputs[2530:2550] = 2
inputs[2740:2760] = 1
inputs[2850:2870] = 2
inputs[3380:3400] = 1
inputs[3420:3440] = 2

```

```python
input1 = (inputs==0).astype(np.uint8)
input2 = (inputs==1).astype(np.uint8)
input3 = (inputs==2).astype(np.uint8)
inputs_spr = np.vstack([input1,input2,input3])
plt.plot(input1)
inputs_spr = cp.asarray(inputs_spr)
```

```python
matfile = './embed_data/roam_real_ica.mat'
signals = scipy.io.loadmat(matfile)
targets = cp.asarray(signals['roam_real_ica'][:,0:3500]) #(7,T)
dt = 0.1
nt = 2
N = 500
alpha = 2* float(cp.random.rand(1))
fb = 1.5 + float(cp.random.rand(1))
g = 1.6
Pz = 0.13
Pgg = 0.1
nn = Force_multi_out_gpu.Reservoir(N=N,p=Pgg,g=g)
nn.get_Jz(7,Pz,fb=fb) #(Nout,pz,fb=1)
nn.add_input(3,0.3,1)
# simtime = cp.arange(0,time_sec,step=dt).reshape(1,-1)
# simtime2 = cp.arange(time_sec,2*time_sec,step=dt).reshape(1,-1)
T = targets.shape[1]
simtime = cp.arange(T)
targets = Force_multi_out_gpu.norm_weight(targets,[5,5,2,2,5,5,5])
[train_out,test_out,weight_train] = nn.internal_train(inputs_spr,targets,dt,alpha,nt,nl=0.1,)
train_out = cp.asnumpy(train_out)
test_out = cp.asnumpy(test_out)
weight_train = cp.asnumpy(weight_train)
simtime = cp.asnumpy(simtime)
targets = cp.asnumpy(targets)
```

```python
plot_range = 3500
plt.figure()
    # print(test_out[0,0:500])
plt.subplot(5,1,1)
plt.plot(simtime.T,targets[0:2,0:plot_range].T,'b',simtime.T,train_out[0:2,0:plot_range].T,'r')
plt.title('training')
plt.subplot(5,1,2)
plt.plot(simtime.T,targets[2:4,0:plot_range].T,'b',simtime.T,train_out[2:4,0:plot_range].T,'r')
plt.title('training 5-7')
plt.subplot(5,1,3)
plt.plot(simtime.T,test_out[0:2,0:plot_range].T,'g')
plt.title('testing1')
plt.subplot(5,1,4)
plt.plot(simtime.T,test_out[2:4,0:plot_range].T,'g')
plt.title('testing2')
plt.subplot(5,1,5)
plt.plot(simtime.T,weight_train[:,0:plot_range].T)
plt.title('weight')
plt.figtext(0.6, 0.01, "g={} , fb={} , Pz={} , Pgg={} \n alpha={}, N={}".format(g,fb,Pz,Pgg,alpha,N), ha="center", fontsize=12, bbox={"facecolor":"orange", "alpha":0.6, "pad":6})
# ct = datetime.datetime.now().strftime("%H_%M_%S")
# filename = './image/intn_worm_' + ct +'.jpeg'
# plt.savefig(filename,dpi=300)
```
